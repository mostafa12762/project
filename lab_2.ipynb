```python
import matplotlib.pyplot as plt

# Sample data
height = [160, 165, 170, 175, 180, 185]
weight = [55, 60, 65, 70, 75, 80]

# Scatter plot
plt.scatter(height, weight, color='blue', marker='*')
plt.title('Height vs Weight Scatter Plot')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Sample dataset
data = pd.DataFrame({
    'Height': [160, 165, 170, 175, 180, 185],
    'Weight': [55, 60, 65, 70, 75, 80],
    'Gender': ['M', 'F', 'M', 'F', 'M', 'F']
})

# Scatter plot colored by Gender
sns.scatterplot(data=data, x='Height', y='Weight', hue='Gender', style='Gender', s=100)
plt.title('Height vs Weight by Gender')
plt.show()
```

Line Plot

```python
import matplotlib.pyplot as plt

# Example: Monthly Sales
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [150, 200, 180, 220, 210, 250]

plt.plot(months, sales, color='blue', marker='o', linestyle='-', linewidth=2)
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)  # adds grid
plt.show()
```

```python
import seaborn as sns
import pandas as pd

data = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [150, 200, 180, 220, 210, 250]
})

sns.lineplot(data=data, x='Month', y='Sales', marker='o', color='green')
plt.title('Monthly Sales Trend with Seaborn')
plt.grid(True)
plt.show()
```

Boxplot

```python
import matplotlib.pyplot as plt

# Example dataset
data = [10, 12, 14, 13, 15, 20, 25, 30, 100]  # Notice 100 is an outlier

# Create box plot
plt.boxplot(data)

# Add title
plt.title("Box Plot Example - Matplotlib")

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example dataset
data = [10, 12, 14, 13, 15, 20, 25, 30, 100]

# Create box plot
sns.boxplot(data=data)

# Add title
plt.title("Box Plot Example - Seaborn")
plt.show()
```

```python
import pandas as pd

# Example DataFrame
df = pd.DataFrame({
    'Temperature': [30, 32, 31, 35, 33, 50, 29],
    'Gas Level': [5, 7, 6, 8, 50, 6, 5],
    'CO2': [400, 405, 410, 420, 415, 425, 430]
})

# Using Seaborn
sns.boxplot(data=df)
plt.title("Box Plot for Multiple Features")
plt.show()
```
